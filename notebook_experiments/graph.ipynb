```python
import matplotlib.pyplot as plt

# Placeholder accuracies for 5 models
models = ['llama3.1-8b-FT-newsmtsc', 'GPT-4o', 'RNN', 'LSTM', 'GRU']
accuracies = [0.7646, 0.6725, 0.42, 0.36, 0.39]  # Replace these with actual values later

# Define a list of colors (one for each bar)
colors = ['skyblue', 'lightgreen', 'salmon', 'lightcoral', 'lightpink']

# Create the bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(models, accuracies, color=colors, edgecolor='black')

# Add titles and labels
plt.title('Model Accuracy Comparison', fontsize=16)
plt.xlabel('Models', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.ylim(0, 1)  # Assuming accuracy values range between 0 and 1

# Annotate bars with accuracy values
for i, accuracy in enumerate(accuracies):
    plt.text(i, accuracy + 0.02, f'{accuracy:.2f}', ha='center', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
